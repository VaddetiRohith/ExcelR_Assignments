```python
x = df.iloc[:, 1:]
y = df['Type']
```

### Feature Scaling 

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
x
```

```python
# Import PCA class
pca = PCA()
pcs = pca.fit_transform(x)
pcs
```

```python
# PCA Components matrix or convariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
pca.explained_variance_ratio_
```

```python
# Cummulative variance of each PCA
np.cumsum(pca.explained_variance_ratio_)
```

```python
# Scree plot or elbow plot

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Features')
plt.ylabel('Explained Variance')
plt.grid()
```

```python
pca=PCA(n_components=3)
pcs = pca.fit_transform(x)
pcs
```

```python
data = pd.concat([y, pd.DataFrame(pcs, columns=['PC1', 'PC2', 'PC3'])],axis=1)
data
```

```python
plt.rcParams['figure.figsize']=(12,7)
plt.rcParams['figure.dpi'] = 200
```

```python
# Visualization of PCAs
sns.scatterplot(data= data)
```

```python
sns.scatterplot(data= data, x= 'PC1',y='PC2', hue='Type')
```

## Checking with other Clustering Algorithms 

### Hierarchical Clustering

```python
lk = hierarchy.linkage(x, method='complete')
dendrogram = hierarchy.dendrogram(lk)
```

```python
hc = AgglomerativeClustering(n_clusters=3)
ylabel = hc.fit_predict(x)
```

```python
df1 = df.copy()
df1['Cluster'] = ylabel
```

```python
df1['Cluster'].value_counts()
```

```python
df1.head()
```

### KMeans

```python
df.head()
```

```python
km = KMeans()
ylabel = km.fit_predict(x)
ylabel
```

```python
wcss = []

for i in range(1,8):
    km = KMeans(n_clusters=i)
    ylabel = km.fit_predict(x)
    wcss.append(km.inertia_)
    
wcss
```

```python
plt.plot(wcss)
```
